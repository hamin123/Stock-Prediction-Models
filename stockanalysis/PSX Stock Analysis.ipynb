```python
HDFC_df["Turnover_value"] = HDFC_df["Close"]*HDFC_df["Turnover"]
```

```python
HDFC_df["Day_Perc_Change"] = HDFC_df["Close"].pct_change()*100
HDFC_df.head()
```

```python
HDFC_df["Day_Perc_Change"].plot(figsize = (12, 6), fontsize = 12)
```

```python
HDFC_df["Day_Perc_Change"].hist(bins = 50, figsize = (10,5)) 
plt.xlabel("Daily returns")
plt.ylabel("Frequency")
plt.show()
#satistics
HDFC_df.Day_Perc_Change.describe()
```

```python
def trend(x):
  
  if x > -0.5 and x <= 0.5:
    return "Slight or No change"
  elif x > 0.5 and x <= 1:
    return "Slight Positive"
  elif x > -1 and x <= -0.5:
    return "Slight Negative"
  elif x > 1 and x <= 3:
    return "Positive"
  elif x > -3 and x <= -1:
    return "Negative"
  elif x > 3 and x <= 7:
    return "Among top gainers"
  elif x > -7 and x <= -3:
    return "Among top losers"
  elif x > 7:
    return "Bull run"
  elif x <= -7:
    return "Bear drop"

HDFC_df["Day_Perc_Change"].count()
HDFC_df["Trend"]= np.zeros(HDFC_df["Day_Perc_Change"].count()+1)
HDFC_df["Trend"]= HDFC_df["Day_Perc_Change"].apply(lambda x:trend(x))
HDFC_df.head()
```

```python
HDFC_df.dropna(axis = 0, inplace = True)
HDFC_df.head()
```

```python
HDFC_pie_data = HDFC_df.groupby("Trend")
pie_label = sorted([i for i in HDFC_df.loc[:, "Trend"].unique()])
plt.pie(HDFC_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```
